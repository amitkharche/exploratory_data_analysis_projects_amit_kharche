# **1. Introduction**
---

Air quality has become a critical focus in urban sustainability due to rising pollution levels from industrialization, vehicular emissions, and population growth. Effective monitoring is essential for assessing pollution exposure, understanding seasonal variations, and enabling data-driven policymaking. This dataset offers a rich source of information for analyzing air pollution through real-time sensor measurements and validated reference analyzer data.

Collected over a one-year period from **March 2004 to February 2005**, the dataset comprises **9358 hourly records** from an **Air Quality Chemical Multisensor Device** deployed at **road level in a polluted Italian city**. It includes sensor responses from **five metal oxide sensors** as well as co-located analyzer readings for critical pollutants like **CO, NOx, NO2, NMHC**, and **Benzene (C6H6)**.

Additionally, the dataset records meteorological conditions such as **temperature**, **relative humidity**, and **absolute humidity**, making it highly suitable for multivariate environmental analysis. Evidence of **sensor drift** and **cross-sensitivity** makes this dataset ideal for exploring **sensor calibration**, **forecasting models**, and **drift compensation** techniques.

Designed strictly for **academic and research use**, this dataset presents valuable opportunities for **EDA**, **predictive modeling**, and **air quality research**, while requiring thoughtful preprocessing due to **missing values coded as -200**.

# **4. Data Acquisition & Description**
---
- This section focuses on acquiring the dataset and understanding its key features.

- The data is sourced from an Air Quality Chemical Multisensor Device deployed in an Italian city. It contains 9358 hourly measurements from March 2004 to February 2005.

- The dataset combines **sensor readings**, **atmospheric conditions**, and **ground truth pollutant levels** from certified analyzers.

- Below is the detailed description of each feature in the dataset:

| Id | Feature             | Description |
|:--:|:--------------------|:------------|
| 01 | Date                | Calendar date of the observation |
| 02 | Time                | Hour of the day (24-hour format) |
| 03 | CO(GT)              | True hourly averaged concentration of Carbon Monoxide in mg/m¬≥ (reference analyzer) |
| 04 | PT08.S1(CO)         | Sensor 1 output (targeted at CO) |
| 05 | NMHC(GT)            | True hourly averaged Non-Methane Hydrocarbons in ¬µg/m¬≥ |
| 06 | C6H6(GT)            | True hourly averaged Benzene concentration in ¬µg/m¬≥ |
| 07 | PT08.S2(NMHC)       | Sensor 2 output (targeted at NMHC) |
| 08 | NOx(GT)             | True hourly averaged Nitrogen Oxides in ppb |
| 09 | PT08.S3(NOx)        | Sensor 3 output (targeted at NOx) |
| 10 | NO2(GT)             | True hourly averaged Nitrogen Dioxide in ¬µg/m¬≥ |
| 11 | PT08.S4(NO2)        | Sensor 4 output (targeted at NO2) |
| 12 | PT08.S5(O3)         | Sensor 5 output (targeted at Ozone) |
| 13 | T                  | Ambient Temperature in ¬∞C |
| 14 | RH                 | Relative Humidity in % |
| 15 | AH                 | Absolute Humidity (derived measure) |

- **Note**: Missing values are represented by `-200`. The dataset captures evidence of sensor drift, cross-sensitivity, and long-term field deployment behavior.

# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

```python
data.isnull().sum().sort_values(ascending= False)
```

```python
### Checking the percentage of NaN values present in each feature

# Making a list of features containing NaN values
features_with_na = [feature for feature in data.columns if data[feature].isnull().sum() > 0]

# Print the feature name and the percentage of missing values
for feature in features_with_na:
    percent_missing = np.round(data[feature].isnull().mean() * 100, 2)
    print(f"{feature}: {percent_missing}% missing values")
```

```python
# Strip whitespace from column names
data.columns = data.columns.str.strip()

# Combine 'Date' and 'Time' columns into a single 'Datetime' column
# Use format='mixed' and dayfirst=True to handle inconsistent formats robustly
data['Datetime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'], format='mixed', dayfirst=True, errors='coerce')

# Drop the original 'Date' and 'Time' columns
data.drop(columns=['Date', 'Time'], inplace=True)

# Display the first 5 rows of the updated DataFrame
data.head()
```

```python
# Drop rows where CO(GT), Temperature (T), or Absolute Humidity (AH) is -200 (i.e., invalid/missing measurements)
data = data[(data['CO(GT)'] != -200) & (data['T'] != -200) & (data['AH'] != -200)]

# Reset index after filtering
data.reset_index(drop=True, inplace=True)

# Check if any -200 values remain in the filtered columns
print("Remaining -200 in CO(GT):", (data['CO(GT)'] == -200).sum())
print("Remaining -200 in T:", (data['T'] == -200).sum())
print("Remaining -200 in AH:", (data['AH'] == -200).sum())

# Preview the cleaned dataset
data.head()
```

**8.1. How does CO concentration vary over time?**

**8.3. What is the monthly trend of NOx concentration?**

**8.10. How does ozone sensor output vary with humidity and temperature?**

```python
sns.scatterplot(data=data, x='RH', y='PT08.S5(O3)', hue='T', palette='coolwarm')
plt.title('PT08.S5(O3) Sensor vs Relative Humidity')
plt.tight_layout()
plt.show()
```

**8.11. How do CO levels differ between daytime and nighttime?**

```python
data['Is_Day'] = data['Hour'].apply(lambda x: 'Day' if 6 <= x <= 18 else 'Night')
sns.boxplot(x='Is_Day', y='CO(GT)', data=data)
plt.title('CO Levels: Day vs Night')
plt.tight_layout()
plt.show()
```

- The boxplot shows that **CO concentrations are slightly higher during the daytime** compared to nighttime, with a **higher median** and upper quartile.
- Both distributions show a wide range with multiple **outliers** above 6 mg/m¬≥, suggesting short-term spikes in carbon monoxide levels likely due to traffic or industrial activity.
- The spread during night hours is slightly narrower, with a **lower interquartile range**, indicating **more stable or less variable CO levels** at night.
- These differences suggest that **human activities such as commuting and daytime operations** significantly influence CO emissions.

**8.12. What is the correlation between all numeric features?**

```python
# Drop non-numeric columns before computing correlation
numeric_data = data.select_dtypes(include=['number'])

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Correlation Matrix of All Numeric Features')
plt.tight_layout()
plt.show()
```

**Observation: Correlation Matrix of All Numeric Features**

- **Strong positive correlations** are observed among:
  - `CO(GT)` and `PT08.S1(CO)` (**0.88**)
  - `C6H6(GT)` and `PT08.S2(NMHC)` (**0.98**)
  - `CO(GT)` and `C6H6(GT)` (**0.93**)
  - These indicate reliable sensor alignment with their targeted gases.

- **Strong negative correlations** are found between:
  - `PT08.S3(NOx)` and `NOx(GT)` (**-0.64**)
  - `CO(GT)` and `PT08.S3(NOx)` (**-0.70**)
  - `T` and `RH` (**-0.57**), showing expected atmospheric behavior where higher temperatures usually reduce relative humidity.

- The `Hour` feature shows **low to moderate positive correlation** with pollutants like `CO(GT)` and `C6H6(GT)`, hinting at daily patterns in emissions likely tied to human activities.

- **Environmental features (T, RH, AH)** are weakly correlated with most pollutants but should not be ignored, as they may influence sensor behavior or pollutant dispersion.

This matrix highlights the importance of **sensor validation, temporal patterns**, and **environmental interactions** in air quality modeling.
