## 8.1. What is the age distribution of users?

````python
sns.set_style('whitegrid')
plt.figure(figsize=(20, 6))
plt.xticks(rotation=45)
sns.countplot(x='age', data=facebook_df)
plt.title("Age Distribution of Facebook Users")
plt.show()
````

<a name = Section82></a>
## 8.2. How are users segmented across defined age groups?
We bucket users into meaningful ranges to compare behavior across life stages.

````python
labels = ['12-14', '15-20','21-30','31-40','41-50', '51-60', '61-70', '71-80','81-90','91-100','101-110','111-120']
facebook_df['age_group'] = pd.cut(facebook_df['age'],
                                  bins=[10,15,20,30,40,50,60,70,80,90,100,110,120],
                                  labels=labels,
                                  include_lowest=True)

plt.figure(figsize=(12, 6))
sns.countplot(x='age_group', data=facebook_df, palette="viridis")
plt.title("Facebook Users Grouped by Age Ranges")
plt.show()
````

````python
facebook_df.groupby('age_group')['age_group'].count()
````

## 8.10. How do users engage through Likes?
Let’s explore likes on the platform: mobile, web, and total.

````python
import matplotlib.pyplot as plt
import seaborn as sns

likes_columns = ['likes', 'mobile_likes', 'www_likes',
                 'likes_received', 'mobile_likes_received', 'www_likes_received']

# Create a 2x3 grid of subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 10))  # Adjust figure size as needed
axes = axes.flatten()  # Flatten the 2D array of axes for easy iteration

# Loop through columns and plot on respective axes
for idx, col in enumerate(likes_columns):
    sns.histplot(facebook_df[col], kde=True, bins=30, ax=axes[idx])
    axes[idx].set_title(f"Distribution of {col.replace('_', ' ').title()}")
    axes[idx].set_xlabel(col.replace('_', ' ').title())
    axes[idx].set_ylabel('Count')

plt.tight_layout()
plt.show()
````

<a name = Section811></a>
## 8.11. What is the platform preference (Mobile vs Web) for liking?

````python
total_likes = facebook_df.likes.count()
mobile_like_missing = facebook_df['mobile_likes'].value_counts().max()
web_like_missing = facebook_df['www_likes'].value_counts().max()

mobile_like_pct = (total_likes - mobile_like_missing) / total_likes * 100
web_like_pct = (total_likes - web_like_missing) / total_likes * 100

print("Percentage of Likes via Mobile:", round(mobile_like_pct, 2))
print("Percentage of Likes via Website:", round(web_like_pct, 2))
````

### **9.1 Conclusion**

This section draws conclusions from the detailed exploratory analysis conducted on the Facebook user dataset:

- The majority of Facebook users fall in the **age group of 15 to 30 years**.
- The **youngest user is 13**, and the **oldest user is 113 years old**.
- **Male users dominate** in terms of count, comprising approximately **59.3%** of the total users.
- **Female users show higher engagement** (likes, friendships) compared to males despite being fewer in number.
- A **large number of users joined Facebook in the last 2 years**, indicating recent growth during the data collection period.
- Most users have **fewer than 500 friends** on Facebook.
- **Male users are more prevalent across most age groups**, except for **ages 51–80**, where **female users slightly outnumber males**.
- Users **under 30 and over 80** tend to have **more friends** than those in middle age.
- **Young females (<30)** initiate more friendships and have more friends than their male counterparts.
- The **maximum tenure is around 9 years**, suggesting the dataset was collected in **2013**, referencing Facebook's launch in **2004**.
- **Tenure is relatively consistent** across both male and female users.
- Across age groups, **female users are more active in liking posts**.
- **Teenagers (13–20)** show the **highest liking activity**.
- This pattern is also reflected in the **likes received**, confirming consistency in user engagement.
- Around **65% of users liked posts via the Mobile App**, while **70% used the Website**.
- **38% of users received likes via Mobile App**, and **63% received them via Website**.
- **Females under 60** tend to initiate more friendships than males.
- **Males over 60** initiate more friendships compared to their female peers.


<a name = Section92></a>
### **9.2 Actionable Insights**

These suggestions are derived from user behavior and can help businesses or marketers tailor their social strategies:

- **Focus content and ads on the 15–30 age group**, as they are the most active demographic.
- **Optimize mobile experiences**, given 65% of users engage via mobile.
- **Encourage engagement from female users**, as they tend to be more interactive and loyal.
- **Consider campaigns for seniors (80+ years)**, who surprisingly show high friend counts and engagement.
- **Newer users (joined in the last 2 years)** are growing fast—offer onboarding campaigns or loyalty rewards.
- **Design separate engagement strategies for teens vs. middle-aged users**, as their behavior differs sharply.
- **Use female influencers in campaigns**, especially targeting users below 30, where they outperform males.
- **Focus social analytics on likes and friendships initiated**, as they are strong indicators of platform activity.
- **Encourage users with low friend counts to connect more**, possibly with “suggested friends” campaigns.

> These insights can help businesses enhance user engagement, personalize marketing, and optimize platform design.
