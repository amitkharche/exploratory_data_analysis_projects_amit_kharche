# **6. Data Cleaning**
---

```python
# Check for Missing Values
missing = data.isnull().sum().sort_values(ascending=False)
percent = round(100 * data.isnull().mean(), 2)
missing_summary = pd.concat([missing, percent], axis=1, keys=['Missing Count', 'Missing %'])
print(missing_summary)
```

```python
# Handle Missing Values

# Impute 'Metascore' with mean
data['Metascore'].fillna(data['Metascore'].mean(), inplace=True)

# Impute 'Revenue (Millions)' with median
data['Revenue (Millions)'].fillna(data['Revenue (Millions)'].median(), inplace=True)
```

```python
# Remove Duplicate Entries
print("Number of duplicate rows:", data.duplicated().sum())
data.drop_duplicates(inplace=True)
```

```python
# Convert Data Types
# Convert 'Runtime (Minutes)' to integer
data['Runtime (Minutes)'] = data['Runtime (Minutes)'].astype(int)
```

```python
# Transform Multi-valued String Columns to Lists

# Convert Genre and Actors into lists
data['Genre'] = data['Genre'].apply(lambda x: x.split(', ') if isinstance(x, str) else [])
data['Actors'] = data['Actors'].apply(lambda x: x.split(', ') if isinstance(x, str) else [])
data['Genre_Count'] = data['Genre'].apply(len)
data['Decade'] = data['Year'].apply(lambda x: str(x)[:3] + '0s')
```

```python
# Generate genre_counts after transforming 'Genre'
genre_list = [genre for sublist in data['Genre'] for genre in sublist]
genre_counts = pd.Series(Counter(genre_list)).astype(int)
```

```python
# Rename Columns for Simplicity (Optional)
data.rename(columns={
    'Runtime (Minutes)': 'Runtime',
    'Revenue (Millions)': 'Revenue'
}, inplace=True)
```

```python
# Final Check for Any Missing Data
print("Any missing values left?", data.isnull().any().any())
```

<a id=section81></a>
**8.1. What is the distribution of IMDb ratings across movies?**

```python
sns.histplot(data['Rating'], bins=20, kde=True)
plt.title('Distribution of IMDb Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

### Observation:
* The **majority of movies** are rated between **6.0 and 7.5**, indicating a strong central concentration of ratings.
* The distribution is **nearly normal** with a slight **left skew**, showing fewer low-rated films.
* Very few entries have a **rating below 4**, suggesting an overall **high-quality selection** in the dataset.


<a id=section82></a>
**8.2. How is the revenue distributed among the movies?**

```python
sns.boxplot(x=data['Revenue'])
plt.title('Boxplot of Revenue')
plt.xlabel('Revenue (Millions)')
plt.show()
```

### Observation:
* The **majority of movies** generate revenue below **200 million**, with a compact interquartile range.
* There are a significant number of **outliers**, with some movies earning up to **900+ million**, showing high variability.
* The data is **positively skewed**, indicating that a few blockbuster hits pull the average revenue upward.



<a id=section83></a>
**8.3. What are the top 10 most frequently occurring genres?**

```python
from collections import Counter

# Generate genre_counts after transforming 'Genre'
genre_list = [genre for sublist in data['Genre'] for genre in sublist]
genre_counts = Counter(genre_list)

pd.Series(genre_counts).nlargest(10).plot(kind='barh', color='skyblue')
plt.title('Top 10 Most Common Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

### Observation:
* **Action, Adventure, Sci-Fi** and **Drama** are the most frequently occurring genres, each appearing in nearly **50 movies**.
* Genre combinations like **Comedy, Drama, Romance** and **Comedy** also have strong representation, reflecting their widespread appeal.
* The top genres highlight a **diverse mix** of action-packed, emotional, and humorous storytelling themes in popular films.

* These combinations suggest a strong audience and critical preference for **heartfelt, meaningful narratives** with compelling plots.
