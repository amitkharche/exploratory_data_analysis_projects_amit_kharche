# **4. Data Acquisition & Description**
---

### Dataset Overview

This section focuses on acquiring the dataset and understanding its key features.

The data is sourced from historical records of **Summer Olympic Games**, covering multiple editions across years and countries. It provides detailed information about individual athlete performances, medals won, and event metadata.

The dataset contains **271,116 records**, each representing a unique athlete-event participation. This enables granular analysis of Olympic trends, country-wise performance, gender participation, and evolution of sports.

The dataset includes **demographic data**, **sports metadata**, and **medal results**.

---

### Feature Description

| Id  | Feature      | Description                                                                  |
|-----|:-------------|:-----------------------------------------------------------------------------|
| 01  | ID           | Unique identifier for each athlete-event entry                               |
| 02  | Name         | Full name of the athlete                                                     |
| 03  | Sex          | Gender of the athlete (`M` for male, `F` for female)                         |
| 04  | Age          | Age of the athlete during the event (may contain missing values)             |
| 05  | Height       | Height of the athlete in centimeters (may contain missing values)            |
| 06  | Weight       | Weight of the athlete in kilograms (may contain missing values)              |
| 07  | Team         | Country or team the athlete represented                                      |
| 08  | NOC          | National Olympic Committee 3-letter country code                             |
| 09  | Games        | Edition of the Olympic Games (e.g., `2008 Summer`)                           |
| 10  | Year         | Year the Olympic Games were held                                             |
| 11  | Season       | Season of the Olympics (`Summer`)                                            |
| 12  | City         | Host city of the Olympic Games                                               |
| 13  | Sport        | General sport category (e.g., Athletics, Swimming)                           |
| 14  | Event        | Specific event within the sport (e.g., Men's 100 metres)                     |
| 15  | Medal        | Medal won (`Gold`, `Silver`, `Bronze`, or `NaN` if no medal won)             |

---

- **Missing values** are present in `Age`, `Height`, `Weight`, and `Medal` columns.
- The dataset allows **longitudinal analysis** of Olympic participation trends, gender balance, sport-wise medal distribution, and nation-specific performance.
- Multiple entries for the same athlete may exist if they competed in more than one event or year.

# **6. Data Cleaning**
---

```python
# Check missing values
print("Missing values per column:")
print(data.isnull().sum())
```

```python
# Fill missing values in 'Country' with 'Unknown'
data['Country'] = data['Country'].fillna('Unknown')
```

```python
initial_shape = data.shape
data = data.drop_duplicates()
print(f"Removed {initial_shape[0] - data.shape[0]} duplicate rows")
```

```python
print(data.dtypes)

# Convert object columns to category for efficiency
categorical_cols = data.select_dtypes(include='object').columns
for col in categorical_cols:
    data[col] = data[col].astype('category')
```

```python
# Strip leading/trailing whitespaces from string columns
for col in data.select_dtypes(include='category').columns:
    data[col] = data[col].str.strip()
```

```python
# Convert to title case
text_columns = ['City', 'Sport', 'Discipline', 'Athlete', 'Country', 'Gender', 'Event', 'Medal']
for col in text_columns:
    data[col] = data[col].str.title()
```

# **8. Exploratory Data Analysis**
---

<a id=section81></a>
**8.1. How has the number of Olympic events changed over the years?**

```python
events_per_year = data.groupby('Year')['Event'].nunique()
events_per_year.plot(kind='line', marker='o', figsize=(10, 5), title='Number of Events Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Unique Events')
plt.grid(True)
plt.show()
```

### Observation:
- The **number of unique Olympic events** has shown a **steady increase** from 1896 to 2000, reflecting the expansion of the Games.  
- There was a **sharp dip during the 1930s and 1940s**, likely due to **World War II disruptions**.  
- The peak occurred in the **early 2000s**, followed by a slight **decline in the most recent editions**.

<a id=section82></a>
**8.2. Which countries have won the most medals overall?**

```python
top_countries = data[data['Medal'].notna()]['Country'].value_counts().head(10)
top_countries.plot(kind='bar', figsize=(10, 5), color='gold', title='Top 10 Countries by Total Medals')
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.xticks(rotation=45)
plt.show()
```

### Observation:
- The **USA** dominates the Olympics with the **highest medal count**, significantly ahead of other nations.  
- Countries like **URS, GBR, and FRA** follow, showing strong historical performance across multiple games.  
- The gap between the **top 1 country and the rest** is notably large, indicating long-term athletic supremacy by the USA.


<a id=section83></a>
**8.3. What is the distribution of medals by type?**

```python
data['Medal'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6, 6), title='Medal Distribution')
plt.ylabel('')
plt.show()
```

### Observation:
- The distribution of medals is **almost evenly split** among **Gold (33.6%)**, **Silver (33.1%)**, and **Bronze (33.3%)**.  
- This indicates a **balanced awarding system** where each medal type is closely comparable in frequency.  
- Such an even spread suggests consistent competition levels across events and time.


<a id=section84></a>
**8.4. How many male and female athletes have participated?**

```python
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution of Athletes')
plt.show()
```

### Observation:
- The chart reveals a **significant gender disparity**, with **male athletes outnumbering female athletes by a large margin**.  
- **Men** account for nearly **three times** the number of **women** in Olympic participation historically.  
- This suggests that **female inclusion in sports** has historically been limited, though trends may be improving in recent years.


<a id=section85></a>
**8.5. How does medal distribution vary by gender?**

```python
sns.countplot(x='Medal', hue='Gender', data=data)
plt.title('Medal Distribution by Gender')
plt.show()
```

### Observation:
- **Male athletes** have consistently secured a **greater number of medals** across all three categories: Gold, Silver, and Bronze.  
- **Female athletes**, while fewer in participation, still show a **balanced medal distribution** among the three types.  
- The gap in total medals highlights a **historical disparity in opportunities and participation** between genders.



<a id=section86></a>
**8.6. Which sports have the highest number of events?**

```python
data['Sport'].value_counts().head(10).plot(kind='barh', figsize=(8, 5), title='Top 10 Sports by Number of Events')
plt.xlabel('Number of Events')
plt.ylabel('Sport')
plt.show()
```

### Observation:
- **Aquatics** and **Athletics** dominate in terms of the **total number of events**, reflecting their diverse range of disciplines.  
- **Rowing** and **Gymnastics** also feature prominently, showcasing their long-standing presence in Olympic programs.  
- Sports like **Sailing**, **Shooting**, and **Wrestling** appear at the lower end, indicating a **narrower event variety** within those categories.

<a id=section87></a>
**8.7. What are the top 10 cities that have hosted the Summer Olympics?**

```python
data['City'].value_counts().head(10).plot(kind='bar', figsize=(10, 5), title='Top 10 Host Cities')
plt.xlabel('City')
plt.ylabel('Number of Events Hosted')
plt.xticks(rotation=45)
plt.show()
```

### Observation:
- **London** stands out as the leading Olympic host city, having conducted the **highest number of events** in history.  
- Other cities like **Athens, Los Angeles, Beijing, and Sydney** also show strong hosting history with over 2,000 events each.  
- This suggests that certain cities serve as **recurring Olympic hubs**, likely due to superior infrastructure and global readiness.

<a id=section88></a>
**8.8. Which years had the highest number of athlete participations?**

```python
participation_count = data.groupby('Year')['Athlete'].nunique()
participation_count.plot(kind='line', marker='o', figsize=(10, 5), title='Unique Athlete Participation Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Unique Athletes')
plt.grid(True)
plt.show()
```

### Observation:
- The number of **unique athletes** has shown a **consistent upward trend**, especially from the **1970s onward**.  
- Periodic **declines between 1920 and 1940** likely correspond to **global events like World Wars** impacting participation.  
- Post-1980, the rise in athletes reflects growing **global inclusivity, expanded events, and broader country representation**.


<a id=section89></a>
**8.9. What is the sport-wise distribution of medals for men vs women?**

```python
medal_gender_sport = data[data['Medal'].notna()]
plt.figure(figsize=(12, 6))
sns.countplot(y='Sport', hue='Gender', data=medal_gender_sport, order=medal_gender_sport['Sport'].value_counts().index[:10])
plt.title('Top 10 Sports - Medal Distribution by Gender')
plt.show()
```

### Observation:
- **Aquatics** and **Athletics** top the chart in total medals, with **men significantly outnumbering women** across most sports.  
- Sports like **Gymnastics** and **Rowing** show **stronger female representation**, indicating more inclusive event structures.  
- **Wrestling**, **Shooting**, and **Sailing** reveal a **notable gender gap**, highlighting opportunities to improve women’s participation in these areas.


<a id=section810></a>
**8.10. How does the number of disciplines evolve over decades?**

```python
data['Decade'] = (data['Year'] // 10) * 10
discipline_by_decade = data.groupby('Decade')['Discipline'].nunique()
discipline_by_decade.plot(kind='bar', figsize=(10, 5), color='mediumseagreen', title='Unique Disciplines per Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Disciplines')
plt.show()
```

### Observation:

- The number of **unique Olympic disciplines** has generally **increased over time**, highlighting the diversification of sports.  
- The **2000s witnessed the highest count**, peaking at over 40 disciplines, followed closely by the 2010s.  
- Periods like the **1910s and 1940s** show slight drops, likely influenced by **global disruptions** such as world wars.

<a id=section811></a>
**8.11. Which athletes have won the most medals?**

```python
top_athletes = data[data['Medal'].notna()]['Athlete'].value_counts().head(10)
top_athletes.plot(kind='barh', figsize=(10, 5), title='Top 10 Medal-Winning Athletes')
plt.xlabel('Total Medals')
plt.ylabel('Athlete')
plt.show()
```

### Observation:
- **Michael Phelps** leads as the **most decorated Olympic athlete**, with an unparalleled total medal count.  
- **Larisa Latynina** and **Nikolay Andrianov** follow closely, showing the dominance of Soviet-era athletes in the Games.  
- The list includes a balanced mix of **gymnasts, swimmers, and fencers**, reflecting excellence across multiple disciplines.


<a id=section812></a>
**8.12. How many medals were won each decade?**

```python
medals_by_decade = data[data['Medal'].notna()].groupby('Decade').size()
medals_by_decade.plot(kind='bar', figsize=(10, 5), title='Medals Won by Decade', color='coral')
plt.xlabel('Decade')
plt.ylabel('Number of Medals')
plt.show()
```

### Observation:

- The total **number of medals awarded** has shown a **steady rise over the decades**, with a significant **peak in the 2000s**.  
- Lower counts in the **1910s, 1940s, and early decades** align with global conflicts and fewer hosted games.  
- The spike in the **2000s** reflects a period of **expanded participation, more events, and global inclusivity** in the Olympics.

<a id=section813></a>
**8.13. Which sports have yielded the highest number of Gold medals?**

```python
# Filter only Gold medal entries
gold_medals = data[data['Medal'].str.title() == 'Gold']  # Ensure proper casing

# Count gold medals by sport
top_sports_gold = gold_medals['Sport'].value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_sports_gold.values, y=top_sports_gold.index, color='gold')
plt.title('Top 10 Sports by Gold Medals')
plt.xlabel('Gold Medal Count')
plt.ylabel('Sport')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Observation
- **Aquatics** leads all sports in terms of **Gold medal count**, followed closely by **Athletics**, reflecting their prominence in Olympic history.  
- **Rowing** and **Gymnastics** also show high gold medal numbers, indicating their **event variety and competitive participation**.  
- Sports like **Shooting**, **Wrestling**, and **Sailing** have fewer gold medals but maintain a **consistent Olympic presence**.:

<a id=section814></a>
**8.14. How does medal count vary across top 5 countries and 3 medal types?**

```python
top5 = data[data['Country'].isin(data['Country'].value_counts().head(5).index)]
sns.countplot(x='Country', hue='Medal', data=top5)
plt.title('Medal Count by Top 5 Countries')
plt.show()
```

### Observation:
- The **USA leads decisively** in all medal categories, especially in **Gold**, showcasing long-term Olympic dominance.  
- **URS (former Soviet Union)** holds second in golds and silvers, underscoring its strong historical performance.  
- Countries like **GBR, GER, and FRA** show a more balanced distribution across **Gold, Silver, and Bronze** medals.

<a id=section815></a>
**8.15. Which Olympic cities have hosted the most medal-winning events?**

```python
# Filter only medal-winning rows
medal_data = data[data['Medal'].notna()]

# Count number of medal-winning events by city
top_cities = medal_data['City'].value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='magma')
plt.title('Top 10 Cities by Medal-Winning Events')
plt.xlabel('Number of Medal-Winning Events')
plt.ylabel('City')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Observation:
- **London** hosted the **highest number of medal-winning events**, significantly ahead of all other cities.  
- Cities like **Athens, Los Angeles, Beijing, and Sydney** also played major roles in Olympic history, with over 2,000 medal events each.  
- The distribution suggests these cities have been **frequent and impactful hosts**, reflecting their infrastructure and historical importance in the Games.
