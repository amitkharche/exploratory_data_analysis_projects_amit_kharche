# **1. Introduction**
---
The **Indian Premier League (IPL)** is one of the most celebrated and commercially successful **T20 cricket leagues** in the world, drawing **global viewership** and featuring the **top international and domestic players**. Since its launch in **2008**, the league has evolved into a **data-rich ecosystem**, offering deep insights into **team strategies**, **player performances**, and **match dynamics**.

Each IPL season generates a vast amount of structured data, from **toss decisions**, **match outcomes**, and **player stats**, to **ball-by-ball delivery events**. This makes the IPL an ideal case study for **data-driven storytelling**, **exploratory data analysis (EDA)**, and **sports analytics**.

This project focuses on performing an in-depth **Exploratory Data Analysis (EDA)** using two key datasets:
- `ipl_matches.csv` — containing **match-level data**
- `ipl_deliveries.csv` — containing **ball-by-ball delivery data**

By merging and analyzing these datasets across multiple seasons, we aim to uncover **hidden patterns**, **performance trends**, and **strategic insights** that define the IPL’s competitive structure. The analysis not only enriches our understanding of the game but also provides a solid foundation for **predictive modeling**, **player benchmarking**, and **tactical simulations** in future phases.

# **2. Problem Statement**
---
The primary objective of this project is to perform a comprehensive **Exploratory Data Analysis (EDA)** on **IPL match-level** and **delivery-level data** to uncover **hidden patterns** and derive **actionable insights**. <a id=section401></a>
### 6.1 Data Preprocessing

```python
# Set 'id' as index in matches dataset
merge_df.set_index('id', inplace=True)
```

```python
# Fill missing values in 'city' column with 'Dubai'
merge_df['city'] = merge_df['city'].fillna('Dubai')
```

```python
# Drop 'umpire3' column due to excessive missing values
merge_df.drop(columns=['umpire3'], inplace=True)
```

```python
# Replace 'Rising Pune Supergiant' with 'Rising Pune Supergiants' in relevant columns
merge_df['team1'] = merge_df['team1'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
merge_df['team2'] = merge_df['team2'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
merge_df['winner'] = merge_df['winner'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
merge_df['batting_team'] = merge_df['batting_team'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
merge_df['bowling_team'] = merge_df['bowling_team'].replace('Rising Pune Supergiant', 'Rising Pune Supergiants')
```

```python
# Replace city and venue names
merge_df['city'] = merge_df['city'].replace('Bangalore', 'Bengaluru')
merge_df['city'] = merge_df['city'].replace('Chandigarh', 'Mohali')
merge_df['venue'] = merge_df['venue'].replace( 'Punjab Cricket Association Stadium, Mohali','Punjab Cricket Association IS Bindra Stadium, Mohali')
```

# **8. Exploratory Data Analysis**

<a id=section81></a>
**8.1. Which teams have the most wins across all seasons?**

```python
top_winners = merge_df.drop_duplicates(subset='match_id')['winner'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_winners.values, y=top_winners.index, palette='viridis')
plt.title('Top 10 Teams with Most Wins', fontsize=16)
plt.xlabel('Wins', fontsize=14)
plt.ylabel('Team', fontsize=14)
plt.tight_layout()
plt.show()
```

### Observation:
- **Mumbai Indians** lead the chart with the **highest number of wins**, followed closely by **Chennai Super Kings** and **Kolkata Knight Riders**.
- The top **4–5 teams** show a **competitive win count**, indicating consistent performance across seasons.
- Teams like **Deccan Chargers** and **Rising Pune Supergiants** have **fewer wins**, likely due to **shorter participation span** in the IPL.

<a id=section82></a>
**8.2. What is the distribution of victory margins (runs vs. wickets)?**

```python
match_level_df = merge_df.drop_duplicates(subset='match_id')
plt.figure(figsize=(10, 5))
sns.histplot(match_level_df['win_by_runs'], color='blue', label='Win by Runs', kde=True)
sns.histplot(match_level_df['win_by_wickets'], color='green', label='Win by Wickets', kde=True)
plt.legend()
plt.title('Distribution of Victory Margins')
plt.xlabel('Margin')
plt.tight_layout()
plt.show()
```

### Observation:
- The distribution shows most matches are won by **small margins**, with **wins by wickets** peaking around **4–6 wickets** and **wins by runs** clustering below **30 runs**.
- **Wins by wickets** are more frequent, indicating **chasing teams often succeed**.
- A **long tail in win by runs** suggests occasional **dominant victories** by teams batting first.


<a id=section83></a>
**8.3. Which venues hosted the most IPL matches?**

```python
venue_counts = match_level_df['venue'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=venue_counts.values, y=venue_counts.index, palette='Set3')
plt.title('Top 10 IPL Venues by Unique Match Count', fontsize=16)
plt.xlabel('Number of Matches', fontsize=14)
plt.ylabel('Venue', fontsize=14)
plt.tight_layout()
plt.show()
```

### Observation:
- **M Chinnaswamy Stadium**, **Eden Gardens**, and **Feroz Shah Kotla** have hosted the **highest number of IPL matches**, indicating their importance in the tournament.
- These venues are consistently used due to **large fanbases**, **central locations**, and **franchise dominance**.
- Venues like **Dr DY Patil Sports Academy** and **Maharashtra Cricket Association Stadium** appear less frequently, showing **limited or temporary usage**.


<a id=section84></a>
**8.4. What are the toss decision trends: bat or field?**

```python
toss_decision_counts = match_level_df['toss_decision'].value_counts()
plt.figure(figsize=(6, 6))
toss_decision_counts.plot.pie(autopct='%1.1f%%', colors=['#66b3ff', '#99ff99'])
plt.title('Toss Decision Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

### Observation:
- A majority of teams (~**59.3%**) prefer to **field first** after winning the toss.
- Only around **40.7%** choose to **bat first**, reflecting a common **chasing strategy** in T20 cricket.
- This trend highlights teams’ confidence in **pacing a chase** rather than setting a target under pressure.

<a id=section85></a>
**8.5. How often does the toss winner win the match?**

```python
match_level_df['toss_match_win'] = match_level_df['toss_winner'] == match_level_df['winner']
toss_match_win_counts = match_level_df['toss_match_win'].value_counts()
plt.figure(figsize=(6, 5))
sns.barplot(x=toss_match_win_counts.index.map({True: 'Yes', False: 'No'}), 
            y=toss_match_win_counts.values, palette='coolwarm')
plt.title('Did Toss Winner Also Win the Match?')
plt.xlabel('Toss Winner = Match Winner')
plt.ylabel('Match Count')
plt.tight_layout()
plt.show()
```

### Observation:
- The **toss winner also won the match** in nearly **50% of cases**, indicating only a **slight edge or no significant advantage**.
- The counts are **almost equal**, suggesting that **toss is not a decisive factor** in determining the final match outcome.
- This highlights the importance of **in-game strategies and execution** over the initial advantage of the toss.


<a id=section86></a>

### Observation:

- The list includes a mix of **openers and finishers**, such as **Rohit Sharma**, **Chris Gayle**, and **MS Dhoni**, highlighting diverse scoring roles.

<a id=section88></a>
**8.8. Which teams perform better batting first vs chasing?**

```python
match_outcomes = match_level_df[['batting_team', 'win_by_runs', 'win_by_wickets', 'winner']].copy()
match_outcomes['batting_first'] = match_outcomes['win_by_runs'] > 0
bat_first_win = match_outcomes[match_outcomes['batting_first']]['winner'].value_counts()
bat_second_win = match_outcomes[~match_outcomes['batting_first']]['winner'].value_counts()
bat_win_compare = pd.DataFrame({'Bat First Wins': bat_first_win, 'Chase Wins': bat_second_win}).fillna(0).astype(int)
bat_win_compare.head(10).plot(kind='bar', figsize=(12,6), colormap='Accent')
plt.title('Bat First vs Chase Wins (Top Teams)')
plt.xlabel('Team')
plt.ylabel('Win Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Observation:
- Teams like **Mumbai Indians** and **Chennai Super Kings** have a **balanced record**, performing well both while setting and chasing targets.
- **Kolkata Knight Riders** and **Delhi Daredevils** show a **strong chasing preference**, with significantly more wins while chasing.
- Most franchises demonstrate **better performance chasing**, reaffirming the T20 trend of **target pacing advantage** in second innings.

<a id=section811></a>
**8.11. What is the distribution of extras (wides, no-balls)?**

```python
extras_series = pd.Series({
    'Wides': merge_df['wide_runs'].sum(),
    'No Balls': merge_df['noball_runs'].sum()
})
plt.figure(figsize=(6, 5))
sns.barplot(x=extras_series.index, y=extras_series.values, palette=['#ff9999','#66b3ff'])
plt.title('Total Wides vs No Balls')
plt.ylabel('Extras Count')
plt.tight_layout()
plt.show()
```

- The low number of no balls indicates **better front-foot discipline**, but wides continue to contribute heavily to **extra runs**.



<a id=section812></a>
**8.12. What are the most common types of dismissals?**

```python
dismissal_counts = merge_df['dismissal_kind'].value_counts()
plt.figure(figsize=(10, 5))
sns.barplot(x=dismissal_counts.index, y=dismissal_counts.values, palette='cubehelix')
plt.title('Types of Dismissals in IPL')
plt.xlabel('Dismissal Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Rare dismissals like **hit wicket**, **retired hurt**, and **obstructing the field** occur infrequently, showcasing their **exceptional nature** in match scenarios.



<a id=section813></a>
**8.13. Which batting pairs have the highest total partnerships?**

```python
# Group by batsman and non-striker to calculate partnership runs
partnerships = merge_df.groupby(['match_id', 'batsman', 'non_striker'])['batsman_runs'].sum().reset_index()
top_partners = partnerships.groupby(['batsman', 'non_striker'])['batsman_runs'].sum().sort_values(ascending=False).head(10)

# Plot
top_partners = top_partners.reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='batsman_runs', y=top_partners['batsman'] + ' & ' + top_partners['non_striker'], data=top_partners, palette='Spectral')
plt.title('Top 10 Batting Partnerships by Runs')
plt.xlabel('Total Partnership Runs')
plt.ylabel('Batting Pair')
plt.tight_layout()
plt.show()
```

### Observation:

```python
# Calculate total runs conceded and total balls bowled
bowler_stats = merge_df.groupby('bowler').agg(
    runs_conceded=('total_runs', 'sum'),
    balls_bowled=('ball', 'count')
).reset_index()

# Filter for bowlers with at least 500 balls
bowler_stats = bowler_stats[bowler_stats['balls_bowled'] >= 500]
bowler_stats['economy'] = (bowler_stats['runs_conceded'] / bowler_stats['balls_bowled']) * 6
top_economy = bowler_stats.sort_values(by='economy').head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='economy', y='bowler', data=top_economy, palette='BuGn_r')
plt.title('Top 10 Most Economical Bowlers (Min 500 Balls)')
plt.xlabel('Economy Rate')
plt.ylabel('Bowler')
plt.tight_layout()
plt.show()
```

**8.15. How has the average score changed over IPL seasons?**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure seaborn style
sns.set(style="whitegrid")

# Drop duplicate matches for season-level aggregation
match_level_df = merge_df.drop_duplicates(subset='match_id').copy()

# Convert 'date' column to datetime (mixed format, day-first)
match_level_df['season'] = pd.to_datetime(match_level_df['date'], format='mixed', dayfirst=True, errors='coerce')

# Filter out rows where parsing failed
match_level_df = match_level_df.dropna(subset=['season'])

# Extract year
match_level_df['year'] = match_level_df['season'].dt.year

# Total runs per match from delivery-level data
match_runs = merge_df.groupby('match_id')['total_runs'].sum().reset_index()

# Merge with year info
match_scores = match_runs.merge(match_level_df[['match_id', 'year']], on='match_id', how='left')

# Average score per season
avg_score_by_season = match_scores.groupby('year')['total_runs'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_score_by_season, x='year', y='total_runs', marker='o', color='darkorange', linewidth=2.5)
plt.title('Average IPL Match Score per Season', fontsize=16)
plt.xlabel('Season', fontsize=14)
plt.ylabel('Average Total Runs', fontsize=14)
plt.xticks(avg_score_by_season['year'])
plt.tight_layout()
plt.show()
```

### Observation:
- The **average total runs per match** have shown a **steady upward trend**, especially from **2016 to 2018**, with 2018 recording the **highest scoring average**.
- Early seasons like **2009 and 2011** saw lower average scores, likely due to **conservative batting approaches** or **bowler-friendly pitches**.
- The rise in scoring reflects **evolving strategies**, **aggressive power-hitting**, and **batting-friendly rules** introduced in recent IPL editions.
